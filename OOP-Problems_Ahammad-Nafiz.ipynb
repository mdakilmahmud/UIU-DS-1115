# Problem 1

```python
# 1
class Book:
    def __init__(self, title, author, copies):
        self.title = title
        self.author = author
        self.copies = copies

    def __str__(self):
        return f"{self.title} by {self.author} ({self.copies} copies available)"

# 2
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# 3
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
        self.books_borrowed = []

    def borrow_book(self, book):
        if book.copies > 0:
            book.copies -= 1
            self.books_borrowed.append(book)
            print(f"{self.name} borrowed '{book.title}'")
        else:
            print(f"Sorry, '{book.title}' is not available.")

    def return_book(self, book):
        if book in self.books_borrowed:
            book.copies += 1
            self.books_borrowed.remove(book)
            print(f"{self.name} returned '{book.title}'")
        else:
            print(f"{self.name} didn't borrow '{book.title}'")

# 4
class Librarian(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def add_book(self, library, book):
        library.append(book)
        print(f"{self.name} added '{book.title}' to the library.")

    def remove_book(self, library, book):
        if book in library:
            library.remove(book)
            print(f"{self.name} removed '{book.title}' from the library.")
        else:
            print(f"'{book.title}' not found in the library.")


library = []

lib = Librarian("Mr. Khan", 40, 1001)
book1 = Book("Python 101", "Guido", 3)
book2 = Book("Data Science", "Hadley", 2)

lib.add_book(library, book1)
lib.add_book(library, book2)

student = Student("Akil", 22, 1234)
student.borrow_book(book1)
student.borrow_book(book2)

student.return_book(book1)

student.display_info()
```

# Problem 2

```python
# 1
class Product:
    def __init__(self,product_id:int,name:str,price:float):
        self.__product_id=product_id
        self.name=name
        self.__price=price

    def get_price(self):
        return self.__price
    
    def get_product_id(self):
        return self.__product_id
    
    def display_info(self):
        print(f"Product Id: {self.__product_id}, Name: {self.name}, Price: {self.__price} TK")

# 2
class Customer:
    def __init__(self,customer_id: int,name:str):
        self.__customer_id=customer_id
        self.name=name
        self.cart=[]

    def get_customer_id(self):
        return self.__customer_id
    
    def add_to_cart(self,product):
        self.cart.append(product)
        return f"Items in cart: {self.cart}"
    
    def remove_from_cart(self,product):
        if product in self.cart:
            self.cart.remove(product)
            print(f"Removed from cart: {product.name}")
        else:
            print(f"{product.name} is not in the cart.")

    def checkout(self):
        if not self.cart:
            print("Cart is empty.")
            return

        print(f"Checkout for {self.name}:")
        total = 0
        for product in self.cart:
            product.display_info()
            total += product.get_price()
        print(f"Total: ${total:.2f}")
        self.cart.clear()

    def display_info(self):
        print(f"Customer ID: {self.__customer_id}, Name: {self.name}")

# 3
class DiscountedProduct(Product):
    def __init__(self,product_id:int,name:str,price:float,discount_percent:float):
        super().__init__(product_id,name,price)
        self.discount_percent=discount_percent

    def get_price(self):
        return super().get_price()*(1-self.discount_percent/100)
    
    def display_info(self):
        print(f"[DISCOUNTED] {self.name} - ${self.get_price():.2f} (Discount: {self.discount_percent}%)")

# 4
class ElectronicsProduct(Product):
    def __init__(self,product_id:int,name:str,price:int,warranty_years:int):
        super().__init__(product_id,name,price)
        self.warranty_years=warranty_years

    def display_info(self):
        print(f"[ELECTRONICS] {self.name} - ${self.get_price():.2f}, Warranty: {self.warranty_years} years")

#5   
class ClothingProduct(Product):
    def __init__(self,product_id:int,name:str,price:int,size:str):
        super().__init__(product_id,name,price)
        self.size=size
    
    def display_info(self):
        print(f"[CLOTHING] {self.name} - ${self.get_price():.2f}, Size: {self.size}")

p1 = Product(1, "Mouse", 500)
p2 = DiscountedProduct(2, "Keyboard", 1200, 10)
p3 = ElectronicsProduct(3, "Monitor", 15000, 4)
p4 = ClothingProduct(4, "T-Shirt", 800, "M")

cust = Customer(101, "Akil")

cust.add_to_cart(p1)
cust.add_to_cart(p2)
cust.add_to_cart(p3)
cust.add_to_cart(p4)

cust.remove_from_cart(p1)

cust.checkout()
```

# Problem 3

```python
# 1
class User:
    def __init__(self,user_id:int,username:str):
        self.__user_id=user_id
        self.username=username
        self.posts=[]

    def get_user_id(self):
        return self.__user_id
    
    def make_post(self,post_content):
        self.posts.append(post_content)
        print(f"{self.username} posted: {post_content}")

    def display_posts(self):
        if len(self.posts)==0:
            print(f"{self.username} has no post yet!")
        else:
            print(f"Posts by {self.username}")

# 2    
class AdminUser(User):
    def __init__(self,user_id:int,username:str,admin_level:int):
        super().__init__(user_id,username)
        self.__admin_level=admin_level

    def get_admin_level(self):
        return self.__admin_level
    
    def make_post(self,post_content):
        self.posts.append(f"[Admin] {post_content}")
        print(f"{self.username} (Admin) Posted: {post_content}")

    def remove_post(self,user,post_content):
        if post_content in user.posts:
            user.posts.remove(post_content)
            print(f"{self.username} removed the post: {post_content}")
        else:
            print(f"Post not found in {user.username}'s posts.")
        
# 3
class Comment:
    def __init__(self,comment_id:int,user,post_content,comment_content):
        self.comment_id=comment_id
        self.user=user
        self.post_content=post_content
        self.comment_content=comment_content

    def display_comment(self):
        print(f"Comment by {self.user} on '{self.post_content}': {self.comment_content}")

# 4
class RegularUser(User):
    def __init__(self, user_id, username):
        super().__init__(user_id, username)
        self.account_type = "Regular"

# 5
class PremiumUser(User):
    def __init__(self, user_id, username):
        super().__init__(user_id, username)
        self.account_type = "Premium"

user1 = RegularUser(1, "akil")
user2 = PremiumUser(2, "data_wizard")
admin = AdminUser(0, "admin_boss", 5)

# Users making posts
user1.make_post("Just started learning Python!")
user2.make_post("Premium feature unlocked.")
admin.make_post("Please follow community guidelines.")

# Displaying posts
user1.display_posts()
user2.display_posts()
admin.display_posts()

# Admin removing a post
admin.remove_post(user1, "Just started learning Python!")
user1.display_posts()

# Adding and displaying a comment
comment1 = Comment(101, user2, "Please follow community guidelines.", "Thanks for the reminder!")
comment1.display_comment()
```
